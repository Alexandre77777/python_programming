# Практическая работа №1. Парадокс Монти Холла

Парадокс Монти Холла - это известная задача из теории вероятностей, решение которой на первый взгляд противоречит здравому смыслу. Она названа в честь ведущего американского телешоу "Let's Make a Deal".

## Правила игры

1. Участнику предлагается выбрать одну из трех дверей.
2. За одной из дверей находится приз, за двумя другими - ничего.
3. После выбора участника ведущий, зная расположение приза, открывает одну из оставшихся дверей, за которой ничего нет.
4. Затем ведущий спрашивает участника, не хочет ли он изменить свой выбор.

image.png

## Вопрос

- Увеличатся ли шансы участника на выигрыш, если он изменит свой первоначальный выбор?

## Предположения

- Математики утверждают, что шансы увеличатся на 60%.
- Интуитивно кажется, что шансы не изменятся.

## Задание

- Напишите программу, которая подтвердит или опровергнет утверждение математиков. Программу необходимо реализовать в виде процедуры c именем `monty_hall(iterations)`, которая принимает на вход количество итераций (проводимых игр) и **выводит в консоль** целевые вероятности:


```
                  Вероятность выигрыша при смене двери: 66,66%
                  Вероятность выигрыша без смены двери: 33,34%

```


---







### Предложение по реализации

Вместо использования `randrange`, можно применить метод `choice` из модуля `random` и список `[0, 0, 1]`. Это позволит более наглядно представить три двери, где 1 - это дверь с призом, а 0 - пустые двери.

**Пример использования:**
```python
import random

doors = [0, 0, 1]  # 0 - пустая дверь, 1 - дверь с призом
chosen_door = random.choice(doors)  # Первоначальный выбор участника
```
