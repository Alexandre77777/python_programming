# **Практическая работа №5: Игра "Поле чудес"**


## **Описание работы:**



В рамках данной работы Вам нужно разработать упрощённую версию игры "Поле чудес", которая запускается в консоли и предназначена для одного игрока. Игра должна быть реализована с использованием модульной структуры и упакована в пакет, состоящий минимум из двух модулей. Список слов для игры загружается из текстового файла с расширением `.txt`, который по умолчанию располагается в пакете с основными модулями игры.

## **Требования к реализации:**

### 1. **Основной функционал:**


   - **Выбор слова:** Игра выбирает случайное слово из списка, загруженного из текстового файла. Обратите внимание, что слова могут быть разной длины.
   - **Отображение слова:** На старте игра отображает слово, заменяя все буквы на символы (например, "■" или "\u25A0").
   - **Счётчик жизней:** Установить счётчик "жизней" в зависимости от выбранного уровня сложности:
     - Лёгкий уровень: 7 жизней.
     - Средний уровень: 5 жизней.
     - Сложный уровень: 3 жизни.
   - **Ход игрока:** Предложить игроку ввести букву или всё слово целиком.
     - Если буква присутствует в слове, она открывается в соответствующих позициях.
     - Если буква отсутствует, у игрока отнимается одна "жизнь".
     - Игра завершается, если игрок угадал всё слово, ввёл его целиком правильно или у него закончились "жизни".
   - **Продолжение игры:** После каждого тура предложить игроку сыграть ещё раз или отказаться. При согласии счётчик жизней восстанавливается. Игра продолжается до тех пор, пока не закончатся слова в списке или игрок не решит выйти.



### 2. **Структура проекта:**







   - Игра должна состоять минимум из двух модулей, упакованных в один пакет (можно также создавать дополнительные вложенные пакеты, если видите в этом смысл).
   - Список слов загружается из текстового файла `.txt`, расположенного в пакете с модулями игры.
   - Необходимо создать переменную `record`, отвечающую за количество правильно угаданных слов. После каждого угаданного слова увеличивать её значение на единицу.
   - В конце игры выводить сообщение с количеством угаданных слов. Рекорд должен быть записан в текстовый файл.
   - При последующих запусках игры сравнивать текущий результат с рекордом из файла. Если текущий результат выше, обновлять файл новым рекордом.
   - Пункты, связанные с загрузкой слов и обработкой рекордов, реализовать в отдельном модуле в виде функций.

### 3. **Обработка ошибок:**




   - **Ошибка ввода:**
     - Использовать конструкцию `try-except` для обработки некорректного ввода пользователя (например, ввод чисел вместо букв).
     - При возникновении ошибки вывести пользователю понятное сообщение и предложить повторить ввод.
   - **Ошибки при работе с файлами:**
     - При загрузке списка слов из файла обработать возможные исключения:
       - **Файл не найден:** Использовать `FileNotFoundError` для обработки ситуации, когда файл со словами отсутствует. Вывести сообщение об ошибке и завершить работу программы или предложить ввести путь к файлу вручную.
       - **Ошибка чтения файла:** Обработать возможные ошибки чтения файла (например, недостаточно прав доступа) с помощью соответствующих исключений.
     - При записи рекорда в файл обработать исключения, связанные с записью данных.
   - **Другие возможные ошибки:**
     - Предусмотреть обработку непредвиденных ошибок с помощью общего исключения `Exception` и вывести сообщение о том, что произошла непредвиденная ошибка, с рекомендацией перезапустить игру.
