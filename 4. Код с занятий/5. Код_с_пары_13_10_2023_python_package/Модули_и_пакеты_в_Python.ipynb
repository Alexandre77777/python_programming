# **1. Модуль**



**Модуль** в Python — это просто файл с расширением `.py`, который содержит код на языке Python: функции, классы, переменные и т.д. Вы можете использовать функции и классы из модуля в других файлах, импортируя модуль или его отдельные компоненты.


### **Создание модуля**



Допустим, у нас есть файл `my_module.py` со следующим содержимым:






```python
# my_module.py

def greet(name):
    return f"Привет, {name}!"

def farewell(name):
    return f"До свидания, {name}!"
```

Теперь мы можем использовать этот модуль в другом файле, импортируя его функции:

```python
# main.py

import my_module

print(my_module.greet("Алиса"))
print(my_module.farewell("Вероника"))
```

**Вывод:**

```
Привет, Алиса!
До свидания, Вероника!
```

# **2. Пакеты**



**Пакет** в Python — это каталог (папка), содержащий файлы-модули и файл `__init__.py`. Пакеты позволяют организовать модули в иерархическую структуру, предотвращая конфликты имен и облегчая навигацию по проекту.

### **Создание пакета**





Давайте создадим пакет `my_package` со следующей структурой:

```
my_package/
├── __init__.py
├── math_utils.py
└── string_utils.py
```

- **`__init__.py`** — специальный файл, который обозначает каталог как пакет. Он может быть пустым или содержать код инициализации пакета.

**Содержимое файлов:**

**`my_package/__init__.py`**

```python
# __init__.py

# Импортируем функции из подмодулей для удобства
from .math_utils import add, subtract
from .string_utils import uppercase, lowercase
```

**`my_package/math_utils.py`**

```python
# math_utils.py

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b
```

**`my_package/string_utils.py`**

```python
# string_utils.py

def uppercase(text):
    return text.upper()

def lowercase(text):
    return text.lower()
```

Файл `__init__.py` играет важную роль в инициализации пакета. Он позволяет:

- Определять переменные и функции на уровне пакета.
- Выполнять установку пакета при его импортировании.
- Определять список модулей, доступных при импорте с помощью `__all__`.


**Теперь при импорте с использованием `from my_package import *` будут доступны только указанные модули или функции.**

## **Переменная `__name__`**


`__name__` — специальная переменная в Python, которая содержит имя модуля. Когда модуль запускается как основной файл программы, `__name__` принимает значение `"__main__"`. Это позволяет определять точку входа в программу.

## **Переменная `__all__`:**


1. **Назначение**: `__all__` — это список строк, каждая из которых является именем объекта, который должен быть экспортирован из модуля или пакета при использовании `from module import *`. Это позволяет разработчику контролировать, какие части модуля будут доступны для импорта, и скрыть внутренние детали реализации.

2. **Использование**:
   - Если `__all__` определен в модуле, то `from module import *` импортирует только те имена, которые перечислены в `__all__`.
   - Если `__all__` не определен, то `from module import *` по умолчанию импортирует все имена, которые не начинаются с символа подчеркивания `_`.

3. **Пример**:
   ```python
   # example_module.py
   __all__ = ['public_function', 'PublicClass']

   def public_function():
       pass

   def _private_function():
       pass

   class PublicClass:
       pass

   class _PrivateClass:
       pass
   ```

   В этом примере, если вы выполните `from example_module import *`, то будут импортированы только `public_function` и `PublicClass`. `_private_function` и `_PrivateClass` останутся недоступными, так как они не указаны в `__all__`.

4. **Преимущества**:
   - **Инкапсуляция**: Позволяет скрыть внутренние детали реализации, предоставляя только те функции и классы, которые предназначены для использования извне.
   - **Читаемость**: Упрощает понимание того, какие части модуля или пакета являются его публичным API.

5. **Ограничения**:
   - `__all__` влияет только на поведение `from module import *`. Это не ограничивает доступ к объектам модуля при использовании обычного импорта, например, `import module`.

Таким образом, `__all__` является полезным инструментом для управления экспортом модулей и пакетов, помогая поддерживать чистоту и ясность кода.

## **Абсолютный и относительный импорт**



### **Абсолютный импорт**



Абсолютный импорт использует полный путь к модулю относительно корневого пакета. Он ясен и понятен, но может быть неудобен при изменении структуры пакетов.

**Пример абсолютного импорта:**

```python
# within some_module.py

from my_package.math_utils import add
```

### **Относительный импорт**

Относительный импорт позволяет импортировать модули относительно текущего местоположения, используя точки:

- `.` — текущий пакет.
- `..` — родительский пакет.
- `...` — на уровень выше родительского пакета и т.д.

**Пример относительного импорта:**

```python
# within string_utils.py

from .math_utils import add  # Импортируем из того же пакета
```

**Преимущества относительного импорта:**

- Упрощает импорт внутри пакета.
- Легче поддерживать при реорганизации пакетов.

**Важно:** Относительный импорт возможен только внутри пакетов.

## **Как Python ищет модули и пакеты?**

Когда вы импортируете модуль в Python, интерпретатор выполняет поиск в определенных местах, чтобы найти нужный файл. Этот процесс включает несколько шагов:



1. **Проверка встроенных модулей**: Python сначала проверяет, является ли модуль встроенным. Встроенные модули — это те, которые поставляются вместе с интерпретатором Python, такие как `sys` или `os`.

2. **Поиск в `sys.path`**: Если модуль не встроенный, Python ищет его в списке директорий, указанных в переменной `sys.path`. Этот список включает:
   - Текущую директорию, из которой был запущен скрипт.
   - Директории, указанные в переменной окружения `PYTHONPATH`.
   - Стандартные директории, в которых установлены пакеты Python, такие как `site-packages`.

3. **Файлы и каталоги**: Python ищет файлы с расширением `.py` для модулей и каталоги с файлом `__init__.py` для пакетов. Если файл или каталог найден, он загружается и выполняется.

### **Управление путями поиска**


Вы можете управлять путями поиска модулей и пакетов, изменяя `sys.path`. Это полезно, если вы хотите импортировать модули из нестандартных местоположений. Вот как это можно сделать:

- **Изменение `sys.path` в коде**: Вы можете добавить новый путь в `sys.path` прямо в вашем скрипте. Например:
  ```python
  import sys
  sys.path.append('/path/to/your/module')
  ```

- **Использование переменной окружения `PYTHONPATH`**: Вы можете установить переменную окружения `PYTHONPATH`, чтобы добавить дополнительные директории в `sys.path`. Это делается в командной строке перед запуском вашего скрипта:
  ```bash
  export PYTHONPATH=/path/to/your/module
  ```

- **Создание пакетов**: Если вы создаете собственные пакеты, убедитесь, что структура каталогов соответствует требованиям Python. Каждый пакет должен содержать файл `__init__.py`, даже если он пустой. Это позволяет Python распознавать каталог как пакет.

### **Зачем это нужно?**










Понимание того, как Python ищет модули и пакеты, важно для эффективной организации вашего кода. Это позволяет вам:

- **Избегать конфликтов имен**: Правильная структура пакетов помогает избежать конфликтов имен между модулями.
- **Упрощать повторное использование кода**: Вы можете легко импортировать и использовать код из других проектов или общих библиотек.
- **Управлять зависимостями**: Знание путей поиска помогает вам управлять зависимостями вашего проекта, особенно если вы используете виртуальные среды.

Таким образом, понимание механизма поиска модулей и пакетов в Python и умение управлять путями поиска — это важные навыки для любого разработчика, работающего с этим языком.

# **3. Создание установочного файла пакета (Wheel)**





## **Зачем создавать Wheel-пакеты?**

Когда вы разрабатываете Python-пакеты, особенно те, которые могут быть полезны другим людям или могут быть переиспользованы в разных проектах, возникает необходимость в удобном способе их распространения и установки. Создание установочного файла в формате Wheel (`.whl`) позволяет:

- **Упростить установку пакета**: Wheel-файлы можно установить с помощью `pip`, что делает процесс установки простым и быстрым.
- **Повысить портативность**: Wheel-файлы являются самодостаточными и не зависят от платформы (если содержат только Python-код), что облегчает распространение пакета на разных системах.
- **Стандартизировать распространение**: Формат Wheel является стандартом для пакетов Python, рекомендованным сообществом Python.

### **Где это используется?**


- **Публикация на PyPI**: Если вы хотите поделиться своим пакетом с сообществом Python, вы можете опубликовать его на [Python Package Index (PyPI)](https://pypi.org/). Другие пользователи смогут установить ваш пакет с помощью команды `pip install ваш_пакет`.
- **Развертывание в корпоративной среде**: В компаниях часто создаются внутренние пакеты, которые распространяются среди команд разработки. Wheel-файлы позволяют быстро устанавливать и обновлять такие пакеты.
- **Упрощение установки зависимостей**: При разработке проектов вы можете создавать Wheel-пакеты для своих библиотек и легко устанавливать их как зависимости в других проектах.

Создание Wheel-пакета для вашего проекта делает его более профессиональным и удобным для распространения, что особенно важно при разработке многократно используемых библиотек или приложений.

## **Практический пример: создание Wheel-пакета для `calculator`**



Теперь перейдем к практической части и рассмотрим, как создать установочный файл пакета (Wheel) для нашего пакета `calculator`.

### **Последовательность шагов для создания Wheel-пакета**


Чтобы создать Wheel-пакет для пакета `calculator`, созданного ранее, выполним следующие шаги:

1. **Организуем структуру проекта для публикации.**
2. **Создадим файл `setup.py`.**
3. **Создадим файл `README.md`.**
4. **Создадим Wheel-пакет с помощью `setuptools`.**

#### **1. Организация структуры проекта**

Создадим корневую директорию для нашего проекта:

```
calculator_project/
├── calculator/
│   ├── __init__.py
│   ├── basic/
│   │   ├── __init__.py
│   │   ├── addition.py
│   │   └── subtraction.py
│   └── advanced/
│       ├── __init__.py
│       ├── exponentiation.py
│       └── root.py
├── setup.py
├── README.md
```

- **`calculator/`** — директория с нашим пакетом.
- **`setup.py`** — установочный скрипт для пакета.
- **`README.md`** — файл с описанием пакета.

#### **2. Создание файла `setup.py`**

Файл `setup.py` содержит информацию, необходимую для установки пакета. Вот пример содержимого `setup.py`:



```python
# setup.py

import setuptools

with open("README.md", "r", encoding="utf-8") as fh:
    long_description = fh.read()

setuptools.setup(
    name="calculator-pkg-yourusername",          # Уникальное имя пакета
    version="0.1.0",                             # Версия пакета
    author="Ваше Имя",                           # Ваше имя
    author_email="youremail@example.com",        # Ваш email
    description="Простой калькулятор на Python",
    long_description=long_description,           # Длинное описание из README.md
    long_description_content_type="text/markdown",
    url="https://github.com/yourusername/calculator-pkg",  # Ссылка на репозиторий или сайт
    packages=setuptools.find_packages(),         # Поиск всех пакетов внутри проекта
    classifiers=[                                # Классификаторы
        "Programming Language :: Python :: 3",
        "License :: OSI Approved :: MIT License",  # Укажите вашу лицензию
        "Operating System :: OS Independent",
    ],
    python_requires='>=3.6',                     # Требуемая версия Python
)
```

**Объяснение параметров:**

- **`name`**: Уникальное имя вашего пакета. Если планируете публиковать его на PyPI, убедитесь, что имя не занято.
- **`version`**: Версия пакета. Следует увеличивать при обновлениях.
- **`author`** и **`author_email`**: Информация об авторе пакета.
- **`description`**: Краткое описание пакета.
- **`long_description`**: Длинное описание, обычно из файла `README.md`.
- **`long_description_content_type`**: Формат длинного описания (`text/markdown`, если используете Markdown).
- **`url`**: Ссылка на репозиторий или сайт проекта.
- **`packages`**: Список пакетов, которые нужно включить в дистрибутив. `setuptools.find_packages()` автоматически найдет все пакеты в проекте.
- **`classifiers`**: Метаинформация о пакете (версия Python, лицензия и т.д.).
- **`python_requires`**: Минимальная версия Python, необходимая для работы пакета.

#### **3. Создание файла `README.md`**

Этот файл содержит подробное описание вашего пакета, инструкции по установке и использованию. Пример содержимого:

    ## Использование

#### **4. Создание Wheel-пакета**



Теперь, когда все файлы настроены, можно приступить к созданию Wheel-пакета.

**Установите необходимые инструменты:**

Если у вас еще не установлены `setuptools` и `wheel`, установите их:

```bash
pip install setuptools wheel
```

**Создайте Wheel-пакет:**

В командной строке перейдите в корневую директорию вашего проекта (там, где находится `setup.py`) и выполните команду:

```bash
python setup.py bdist_wheel
```

После выполнения команды будет создана папка `dist/`, в которой появится ваш Wheel-файл. Например:

```
dist/
└── calculator_pkg_yourusername-0.1.0-py3-none-any.whl
```

**Wheel-файл готов к установке или распространению.**

#### **5. Установка созданного пакета (для проверки)**



Вы можете установить пакет локально, чтобы проверить его работоспособность:

```bash
pip install dist/calculator_pkg_yourusername-0.1.0-py3-none-any.whl
```

Теперь вы можете использовать свой пакет из любого места в вашей системе:

```python
from calculator.basic import add

print(add(10, 5))  # Вывод: 15
```


#### **6. (Дополнительно) Публикация пакета на PyPI**

Если вы хотите поделиться своим пакетом с сообществом, вы можете опубликовать его на [PyPI](https://pypi.org/).

**Шаги для публикации:**

1. **Создайте учетную запись на PyPI.**

2. **Установите `twine`:**

   ```bash
   pip install twine
   ```

3. **Загрузите пакет на PyPI:**

   ```bash
   twine upload dist/*
   ```

   При успехе ваш пакет станет доступен для установки:

   ```bash
   pip install calculator-pkg-yourusername
   ```

**Важно:** Перед публикацией убедитесь, что ваш пакет не содержит конфиденциальной информации. Также проверьте соответствие требованиям PyPI.
