# **Работа с GIT в редакторе кода**

### **1. Начальная настройка**

**Установка необходимого ПО:**

- **VS Code**: Это редактор кода, в котором мы будем работать. Установите его, следуя инструкциям на официальном сайте. (Также можно использовать **PyCharm**)
- **Git**: Система контроля версий, которая отслеживает изменения в файлах. Установите Git для вашей операционной системы.


**Настройка аутентификации:**

- Откройте VS Code и нажмите на значок учетной записи в левом нижнем углу (выглядит как силуэт человека).
- Выберите "Войти для синхронизации настроек" и следуйте инструкциям для входа в вашу учетную запись GitHub.
- После входа VS Code получит доступ к вашим репозиториям на GitHub.

**Добавление вспомогательной информации во избежание ошибок:**
  - Откройте GIT Bash
  - Введите в консоль `git config --global user.name "Ваше Имя"` и нажмите `Enter`
  - Затем, введите `git config --global user.email Ваша@почта.ru`  и нажмите `Enter`
  - Для проверки введенных данных введите команду: `git config --list`

### **2. Основные операции с репозиторием**





**Создание нового локального репозитория:**

- Откройте папку проекта в VS Code.
- Нажмите на значок Source Control (`Ctrl+Shift+G`) и выберите "Инициализировать репозиторий". Это превратит папку в Git-проект, и Git начнет отслеживать изменения в файлах.
- После применения изменений Вам будет предложено опубликовать текущую ветку в Github, с созданием соответствующего репозитория на сервере.

image.png

**Клонирование удаленного (онлайн) репозитория:**

- Нажмите `Ctrl+Shift+P` и введите "Git: Клонировать".
- Вставьте URL репозитория (например, `https://github.com/username/repository.git`).
- Выберите локальную папку для клонирования. Это создаст локальную копию репозитория на вашем компьютере и установит связь с удаленной версией на GitHub.

### **3. Работа с изменениями**



Когда вы вносите изменения в проект, важно фиксировать их в виде коммитов. **Коммит** — это сохранение текущего состояния вашего проекта, чтобы вы могли вернуться к нему позже.

**Создание коммита:**

- Внесите изменения в файлы.
- Откройте панель Source Control и добавьте измененные файлы в "индекс" (staging area) с помощью знака `+`.
- Введите сообщение коммита, описывающее изменения, и нажмите `Ctrl+Enter` или галочку для создания коммита. Это сохранит изменения в локальной истории проекта.



{E2DC824B-764A-4524-8331-E1314680340B}.png

После этого вы можете отправить свои изменения на GitHub. Это называется "**push**".

**Отправка изменений на GitHub:**

- Нажмите на кнопку "Синхронизировать изменения" или используйте `Ctrl+Shift+P` и команду "Git: Push", чтобы загрузить ваши локальные коммиты на GitHub.

image.png

Таким образом, ваши локальные изменения станут доступны на удаленной платформе, и другие участники проекта смогут их увидеть.


### **4. Работа с ветками**



Ветки позволяют вам разрабатывать новые функции или исправлять ошибки, не влияя на основную версию проекта.

image.png

**Создание новой ветки:**

- Нажмите на название текущей ветки в строке состояния и выберите "Создать ветку".
- Введите имя новой ветки. Новая ветка создается от текущей и содержит все её файлы.

Рисунок9.png

Ветки помогают поддерживать порядок в проекте и дают возможность экспериментировать без риска сломать что-то важное.

**Переключение между ветками:**

- Нажмите на название текущей ветки и выберите нужную ветку из списка. VS Code обновит файлы в соответствии с выбранной веткой.

Рисунок10.png

Когда работа в ветке завершена, вы можете объединить изменения с основной веткой. Для этого создается Pull Request, где вы можете обсудить изменения с командой и принять решение о включении их в основной проект.


### **5. Работа с Pull Request**



**Pull Request (PR)** — это способ предложить изменения в основной проект. Это как заявка на внесение изменений, где вы и ваши коллеги можете обсудить предлагаемый код, оставить комментарии и решить, следует ли его объединить с основной веткой. PR — важный инструмент для командной работы, так как он позволяет проверять и обсуждать код перед его интеграцией.

image.png

**Создание Pull Request:**

- Откройте панель GitHub и нажмите на значок создания Pull Request.
- Заполните описание и нажмите "Создать Pull Request". В PR можно обсуждать изменения и вносить правки. После проверки PR может быть принят (merged) или отклонен.

### **6. Разрешение конфликтов**



Иногда бывает, что несколько человек одновременно изменяют один и тот же участок кода, и тогда возникают конфликты. VS Code поможет вам увидеть, где именно произошли несовпадения, и выбрать, какие изменения оставить.

**При возникновении конфликтов:**

- Откройте конфликтующий файл в VS Code. Интерфейс покажет конфликтующие части кода.
- Выберите, какие изменения оставить, а какие удалить. После разрешения конфликта сохраните файл и создайте коммит с разрешением конфликта.

### **7. Дополнительные функции**


Используйте историю изменений, чтобы видеть, кто и когда вносил изменения в проект. Это удобно для отслеживания прогресса и анализа кода. Вы также можете сравнивать разные версии файлов, чтобы понять, как проект развивался и почему были внесены те или иные изменения.




**История и сравнение:**

- Щелкните правой кнопкой по файлу и выберите "Открыть историю файла", чтобы увидеть все изменения.
- Для сравнения версий выберите два коммита, нажмите правой кнопкой и выберите "Сравнить". Это поможет организовать командную работу над проектом и отслеживать изменения в коде.

**Полезные команды:**

- `Ctrl+Shift+G` - открыть панель Source Control.
- `Ctrl+Shift+P` - открыть палитру команд.
- `Alt+Enter` - открыть контекстное меню файла.
