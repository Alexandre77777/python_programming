# 2. Макеты и управление компонентами


## 2.1 Cоздание пользовательского интерфейса и организация элементов в окне

Макет в PySimpleGUI – это структура, состоящая из списков и групп элементов управления (виджетов), которые определяют, как элементы будут размещены в окне.

* Объект layout в PySimpleGUI используется для создания пользовательского интерфейса посредством определения макета окна и размещения компонентов на этом макете.

* Макет определяет расположение компонентов на окне и состоит из виджетов PySimpleGUI, таких как кнопки, текстовые поля, метки и другие. Компоненты можно размещать как в горизонтальном, так и в вертикальном направлении.

* Макеты представляют собой списки списков, каждый вложенный список представляет собой группу элементов, которые будут располагаться вместе.
Один из наиболее распространенных макетов в PySimpleGUI - это макет [ [элемент1, элемент2], [элемент3, элемент4] ], где элементы располагаются в окне по горизонтали внутри каждого вложенного списка, а внутри окна элементы располагаются по вертикали.

Пример такого макета:

```python
# Импортируем библиотеку PySimpleGUI, которая используется для создания простых графических интерфейсов пользователя на Python.
import PySimpleGUI as sg

# Определяем макет окна. 'layout' будет списком, содержащим списки с виджетами (элементами интерфейса).
layout = [
    # В первом вложенном списке создаем текстовый элемент 'Text' с подписью и поле для ввода 'InputText'.
    [sg.Text('Введите имя:'), sg.InputText()],  # Здесь пользователь может ввести свое имя.
    # Во втором вложенном списке создаем две кнопки - 'OK' и 'Отмена'.
    [sg.Button('OK'), sg.Button('Отмена')]      # Эти кнопки будут использоваться для подтверждения
                                                 # или отмены введенной информации.
]

# Создаем окно с заголовком 'Пример окна' и предварительно заданным макетом 'layout'.
window = sg.Window('Пример окна', layout)

# Читаем значения, которые были введены или выбраны в окне, и событие, которое было вызвано.
# 'window.read()' ожидает какого-либо действия пользователя, например, нажатия кнопки.
event, values = window.read()

# Закрываем окно после того, как пользователь выполнил действие и данные были считаны.
window.close()

# После вызова 'window.close()', созданное окно закроется.
# Здесь также можно было бы обработать события и значения.
# Например, можно проверить, является ли 'event' строкой 'OK',
# и если это так, можно выполнить какое-либо действие с данными, введенными в 'InputText'.
```

> Этот пример показывает окно с двумя строками: первая строка содержит надпись "Введите имя:", а рядом с ней поле для ввода текста. Вторая строка содержит две кнопки "OK" и "Отмена".


Вы можете использовать и другие типы макетов в PySimpleGUI для более сложной организации элементов в окне. Например, макет [[элемент1, элемент2], [элемент3], [элемент4]] будет располагать первые два элемента в первой строке окна, третий элемент - во второй строке, и четвертый элемент - в третьей строке:

```python
# Импортируем модуль PySimpleGUI для работы с GUI.
import PySimpleGUI as sg

# Определяем макет (layout) окна, где каждый список внутри списка представляет один ряд элементов в окне.
layout = [
    # В первом ряду окна помещаем текстовый элемент и кнопку.
    [sg.Text('Первый элемент'), sg.Button('Кнопка')],
    # Во втором ряду окна размещаем чекбокс (флажок).
    [sg.Checkbox('Второй элемент')],
    # В третьем ряду окна добавляем слайдер, который может изменяться от 0 до 100 и имеет горизонтальную ориентацию.
    [sg.Slider((0, 100), orientation='horizontal')]
]

# Создаем окно с названием 'Пример окна' и используя ранее определенный макет.
window = sg.Window('Пример окна', layout)

# Выполняем чтение событий и значений элементов окна.
# 'event' хранит информацию о том, какой элемент был использован (например, кнопка была нажата).
# 'values' хранит значения элементов (например, положение слайдера или состояние чекбокса).
event, values = window.read()

# Закрываем окно после того, как пользователь сгенерировал событие (например, закрытие окна или нажатие кнопки).
window.close()
```

В следующем примере макет содержит одну метку (`sg.Text('Привет, мир!')`) и две кнопки (`sg.Button('OK')` и `sg.Button('Отмена')`). Метка и кнопки располагаются в вертикальном порядке:

```python
# Импорт библиотеки PySimpleGUI для построения графического интерфейса пользователя
import PySimpleGUI as sg

# Определение макета интерфейса графического пользователя
layout = [
    # Добавление текстового элемента на интерфейс с текстом 'Привет, мир!'
    [sg.Text('Привет, мир!')],
    # Добавление двух кнопок, с надписями 'OK' и 'Отмена'
    [sg.Button('OK'), sg.Button('Отмена')]
]

# Создание окна с использованием заданного макета
# 'Мое окно' - это заголовок окна
window = sg.Window('Мое окно', layout)

# Цикл обработки событий, который будет действовать до тех пор, пока окно не будет закрыто
while True:
    # Метод read() ждёт события (например, нажатие кнопки)
    # и возвращает название события (event) и значения (values), введённые пользователем
    event, values = window.read()

    # Проверка на тип события. Если окно закрыто или нажата кнопка 'Отмена'...
    if event == sg.WINDOW_CLOSED or event == 'Отмена':
        # ...то прервать цикл обработки событий, что приведёт к закрытию окна
        break
    # Если была нажата кнопка 'OK'...
    elif event == 'OK':
        # ...вывести в консоль сообщение 'OK'
        print('OK')

# По завершению цикла (после выхода из него), закрыть окно
window.close()
```

> `layout` возвращает объект окна, который можно использовать для отображения окна на экране. В примере окно создается следующим образом: `window = sg.Window('Мое окно', layout)`.
>
> Последующий цикл обработки событий позволяет программе реагировать на действия пользователя, такие как нажатие на кнопки или закрытие окна.
>
> Обработка каждого события выполняется в блоке условного оператора. В данном случае, если событие - это нажатие на кнопку "OK", будет выведено значение "OK".



Далее создадим макет с несколькими видами элементов управления:

```python
# Импортируем библиотеку PySimpleGUI для создания графического интерфейса пользователя
import PySimpleGUI as sg

# Макет (layout) описывает структуру интерфейса пользователя:
# он содержит список списков, где каждый внутренний список соответствует одной строке элементов в интерфейсе
layout = [
    # Создается текстовая метка и поле для текстового ввода. Они располагаются в одной строке.
    [sg.Text('Введите Ваше имя:'), sg.InputText()],

    # Создается текстовая метка и спиннер (вращающееся числовое поле) для выбора числа в заданном диапазоне.
    # Инициализируется начальным значением 18
    [sg.Text('Введите Ваш возраст:'), sg.Spin([i for i in range(1, 101)], initial_value=18)],

    # Создается текстовая метка и выпадающий список для выбора цвета с предустановленным значением "Зеленый"
    [sg.Text('Выберите ваш любимый цвет:'), sg.Combo(['Красный', 'Зеленый', 'Синий'], default_value='Зеленый')],

    # Создается флажок с предустановленной галочкой
    [sg.Checkbox('Подтверждаю данные', default=True)],

    # Создаются две кнопки: "Ok" для подтверждения введенной информации и "Cancel" для отмены.
    [sg.Button('Ok'), sg.Button('Cancel')]
]

# Создается окно с указанным заголовком и макетом, созданным выше
window = sg.Window('Пример макета PySimpleGUI', layout)

# Основной цикл программы, в котором происходит чтение и обработка событий интерфейса
while True:
    # Происходит захват событий и значений из элементов интерфейса.
    # event - имя события (например, название кнопки), values - словарь введенных значений
    event, values = window.read()

    # Проверка на событие закрытия окна (крестик) или нажатие кнопки "Cancel"
    if event == sg.WIN_CLOSED or event == 'Cancel':
        break  # Выход из цикла, если окно закрыто или нажата кнопка "Cancel"

    # Проверка на событие нажатия кнопки "Ok"
    if event == 'Ok':
        # Функция popup создает модальное диалоговое окно с сообщением
        sg.popup(f"Привет, {values[0]}! Твой возраст: {values[1]}, любимый цвет: {values[2]}", title='Информация')

# После выхода из цикла происходит закрытие окна.
window.close()
```

## 2.2 Управление видимостью, активностью и состоянием компонентов


В PySimpleGUI управление видимостью, активностью и состоянием компонентов осуществляется с помощью методов и функций, предоставляемых библиотекой.

Рассмотрим пример кода, демонстрирующий управление видимостью компонента:

```python
# Импортируем модуль PySimpleGUI для создания GUI
import PySimpleGUI as sg

# Определяем макет окна, состоящий из элементов: чекбокса и текстового поля, которое изначально скрыто
layout = [
    # Создаем строку в макете с чекбоксом
    [sg.Checkbox('Показать текстовый ввод', key='checkbox', enable_events=True)],
    # Создаем строку в макете с текстовым вводом, который по умолчанию не видим
    [sg.Input('Скрытый текстовый ввод', key='input', visible=False)],
    # Создаем строку в макете с кнопкой "Проверить"
    [sg.Button('Проверить')]
]

# Создаем окно с заголовком и макетом, описанным выше
window = sg.Window('Пример управления видимостью').Layout(layout)

# Запускаем цикл обработки событий
while True:
    # Ожидаем событие от окна и получаем введенные значения
    event, values = window.Read()
    # Если была нажата кнопка закрытия окна, выходим из цикла
    if event is None:
        break
    # Если было событие связанное с чекбоксом, обновляем видимость текстового поля
    if event == 'checkbox':
        # Меняем параметр видимости (visible) элемента 'input' в зависимости от состояния чекбокса
        window.Element('input').Update(visible=values['checkbox'])

# После выхода из цикла закрываем окно
window.Close()
```

> В данном примере показано, как при помощи флажка можно управлять видимостью текстового ввода. Когда флажок не отмечен, текстовый ввод скрыт, а когда флажок отмечен, текстовый ввод становится видимым.

Далее рассмотрим пример управления активностью компонента:

```python
# Импортируем модуль PySimpleGUI, который используется для создания графического интерфейса пользователя (GUI).
import PySimpleGUI as sg

# Описываем макет (layout) окна GUI с использованием списков списков виджетов.
layout = [
    # Создаём виджет текста с приглашением ввести имя.
    [sg.Text('Введите имя:')],
    # Создаём виджет для ввода текста, который имеет уникальный ключ 'input' и включенные события для отслеживания изменений.
    [sg.Input(key='input', enable_events=True)],
    # Создаем кнопку 'Сохранить', которая изначально неактивна (disabled=True) и имеет уникальный ключ 'button'.
    [sg.Button('Сохранить', disabled=True, key='button')]
]

# Создаём объект Window, который будет отображать окно с названием 'Пример управления активностью' и использующее ранее определенный макет.
window = sg.Window('Пример управления активностью').Layout(layout)

# Запускаем бесконечный цикл обработки событий для этого окна.
while True:
    # Читаем события и значения элементов интерфейса, ждём их появления (например, нажатия кнопки или ввода текста).
    event, values = window.Read()
    # Если событий нет (например, окно закрыли), прерываем цикл и завершаем программу.
    if event is None:
        break
    # Проверяем, не пустое ли значение в поле ввода (где ключ 'input').
    if values['input'] != '':
        # Если поле ввода не пустое, делаем кнопку 'Сохранить' активной (то есть пользователь может нажать на неё).
        window.Element('button').Update(disabled=False)
    else:
        # Если поле ввода пустое, деактивируем кнопку 'Сохранить', чтобы пользователь не мог на неё нажать.
        window.Element('button').Update(disabled=True)

# По завершению работы с окном закрываем его.
window.Close()
```

> В данном случае кнопка "Сохранить" изначально неактивна. Но, если пользователь введет текст в текстовое поле, то кнопка станет активной. Если же текст будет удален, кнопка снова станет неактивной.

```python
# Импорт модуля PySimpleGUI, который используется для создания графического интерфейса пользователя.
import PySimpleGUI as sg

# Создание макета интерфейса, который будет использоваться для окна.
layout = [
    # Добавляем текстовый элемент (надпись), который приглашает пользователя выбрать опцию.
    [sg.Text('Выберите опцию:')],

    # Добавляем выпадающее меню с двумя опциями и уникальным идентификатором 'option'.
    [sg.OptionMenu(['Опция 1', 'Опция 2'], key='option')],

    # Добавляем кнопку, которую пользователь может нажать, чтобы проверить выбранную опцию.
    [sg.Button('Проверить')]
]

# Создаем окно с заголовком 'Пример управления состоянием' и указанным макетом.
window = sg.Window('Пример управления состоянием').Layout(layout)

# Запускаем бесконечный цикл обработки событий.
while True:
    # Производится чтение событий и значений элементов интерфейса.
    event, values = window.Read()

    # Если событие равно None, это означает, что окно было закрыто пользователем.
    # В этом случае выходим из цикла.
    if event is None:
        break

    # Проверяем значение ключа 'option' из словаря значений, чтобы узнать выбор пользователя.
    if values['option'] == 'Опция 1':
        # Если выбрана 'Опция 1', выводится всплывающее окно с соответствующим сообщением.
        sg.Popup('Вы выбрали опцию 1')
    elif values['option'] == 'Опция 2':
        # Если выбрана 'Опция 2', выводится всплывающее окно с сообщением об этом.
        sg.Popup('Вы выбрали опцию 2')

# После выхода из цикла окно закрывается.
window.Close()
```

> В этом примере есть выпадающий список с двумя опциями. В зависимости от выбранной опции, будет отображаться соответствующее информационное окно.

Все эти примеры демонстрируют возможности PySimpleGUI по управлению видимостью, активностью и состоянием компонентов. Вы можете применить эти подходы в своих собственных проектах, чтобы контролировать поведение графического интерфейса.

## 2.3. Работа с несколькими окнами

Ниже представлен пример создания двух окон, где первое окно позволяет перейти во второе, используя функции:

```python
# Импортируем библиотеку PySimpleGUI для создания графического пользовательского интерфейса (GUI)
import PySimpleGUI as sg

# Определение функции для создания основного окна
def create_main_window():
    # Установить тему для окна
    sg.theme('DarkAmber')

    # Определение компоновки окна, состоящей из виджетов (элементов GUI)
    layout = [
        [sg.Text('Основное окно')],  # Текстовый виджет (надпись)
        [sg.Button('Перейти к окну 2'), sg.Button('Выход')]  # Добавляем две кнопки
    ]

    # Возврат объекта Window с заголовком 'Окно 1' и вышеопределённой компоновкой
    return sg.Window('Окно 1', layout, finalize=True)

# Определение функции для создания второго окна
def create_second_window():
    # Установить другую тему для второго окна
    sg.theme('LightBlue')

    # Определение компоновки второго окна
    layout = [
        [sg.Text('Второе окно')],  # Текстовый виджет
        [sg.Button('Назад'), sg.Button('Выход')]  # Две кнопки для навигации и выхода
    ]

    # Возврат объекта Window с заголовком 'Окно 2' и компоновкой
    return sg.Window('Окно 2', layout, finalize=True)

# Создание первого окна и инициализация переменной window2 значением None
window1, window2 = create_main_window(), None

# Основной цикл сообщений
while True:
    # Чтение событий и значений из всех окон
    window, event, values = sg.read_all_windows()

    # Обработка события закрытия окна или нажатия кнопки 'Выход' в основном окне
    if window == window1 and event in (sg.WIN_CLOSED, 'Выход'):
        break # Прервать цикл, если нужно закрыть основное окно

    # Обработка события закрытия окна или нажатия кнопки 'Выход' во втором окне
    elif window == window2 and event in (sg.WIN_CLOSED, 'Выход'):
        window2.close()  # Закрытие второго окна
        window2 = None   # Сброс ссылки на второе окно

    # Обработка события нажатия кнопки 'Перейти к окну 2' в основном окне
    if window == window1 and event == 'Перейти к окну 2':
        window2 = create_second_window()  # Создать второе окно
        window1.hide()   # Скрыть основное окно

    # Обработка события нажатия кнопки 'Назад' во втором окне
    if window == window2 and event == 'Назад':
        window2.close()  # Закрытие второго окна
        window2 = None   # Сброс ссылки на второе окно
        window1.un_hide() # Показать основное окно снова

# Закрытие основного окна после выхода из цикла
window1.close()
```

> В этом скрипте мы создаем две функции: create_main_window() и create_second_window(), которые возвращают различные окна (sg.Window). Основное окно содержит кнопку для перехода ко второму окну, а второе окно содержит кнопку для возвращения обратно. В главном цикле программы мы обрабатываем события взаимодействия с окнами. Функция sg.read_all_windows() позволяет читать события из всех активных окон, благодаря чему управление многооконным интерфейсом становится простым.

## 3.2 Текстовые поля

Текстовые поля позволяют пользователю вводить текст. Они могут быть однострочными или многострочными.

Пример:

```python
# Импорт библиотеки PySimpleGUI, которая позволяет строить графический интерфейс пользователя.
import PySimpleGUI as sg

# Определение макета окна. Это структура, описывающая элементы интерфейса и их расположение.
layout = [
    # Строка макета с одним текстовым полем для ввода данных пользователем.
    [sg.InputText()],
    # Строка макета с компонентом для вывода текста (консолью вывода программы) заданного размера.
    [sg.Output(size=(30, 5))],
    # Строка макета с кнопкой, нажатие на которую приведет к выполнению какого-либо действия.
    [sg.Button('Ввод')]
]

# Создание окна с использованием ранее определенного макета. Устанавливаем заголовок окна 'Мое окно'.
window = sg.Window('Мое окно', layout)

# Основной цикл программы, в котором происходит чтение событий интерфейса (нажатие кнопок, ввод текста и т.п.)
while True:
    # Чтение событий (действий пользователя) и значений (например, текста, введенного в поля).
    event, values = window.read()

    # Проверка на событие закрытия окна. Если окно закрыто пользователем, выходим из цикла.
    if event == sg.WINDOW_CLOSED:
        break

    # Проверка на событие нажатия кнопки 'Ввод'.
    if event == 'Ввод':
        # Вывод в текстовую область `sg.Output` значения, введенного в первое текстовое поле `InputText()`.
        print(values[0])

# Закрытие окна после выхода из основного цикла программы.
window.close()
```

Рассмотрим следующий пример с реализованной валидацией вводимых данных:

```python
# Импортируем модуль PySimpleGUI для создания GUI
import PySimpleGUI as sg

# Определяем макет окна: текстовую метку, поле ввода и кнопку.
layout = [
    # Создаем текстовую метку с инструкцией для пользователя
    [sg.Text('Введите целое число:')],
    # Добавляем поле для ввода текста с уникальным ключом, чтобы получать доступ к данным
    [sg.Input(key='number_input')],
    # Добавляем кнопку, которая будет отвечать за обработку ввода
    [sg.Button('OK')]
]

# Создаем окно с использованием созданного макета
window = sg.Window('Мое окно', layout)

# Входим в главный цикл обработки событий
while True:
    # Читаем события и значения элементов интерфейса
    event, values = window.read()

    # Определяем реакцию на закрытие окна пользователем
    if event == sg.WINDOW_CLOSED:
        break  # Прерываем цикл, если окно было закрыто

    # Действия при нажатии на кнопку 'OK'
    if event == 'OK':
        # получаем введенное значение из текстового поля по его ключу
        number_input = values['number_input']

        # проверяем, является ли введенное значение числом
        if number_input.isdigit():
            # Если введено число, выводим сообщение о корректном вводе
            sg.popup('Вы ввели целое число!')
        else:
            # Если введено не число, выводим сообщение об ошибке
            sg.popup('Пожалуйста, введите целое число!')

# Закрываем окно после выхода из цикла
window.close()
```

> В этом примере мы создаем окно с текстовым полем для ввода числа и кнопкой 'OK'. При нажатии на кнопку, значение из текстового поля получается с помощью values['number_input']. Затем проверяется, является ли введенное значение целым числом с помощью метода .isdigit(). Если проверка проходит успешно, выводится всплывающее окно с сообщением 'Верное число!'. Если введенное значение не является числом, выводится сообщение 'Пожалуйста, введите число!'.

## 3.3 Выпадающие списки

Выпадающие списки позволяют пользователю выбрать один из предопределенных вариантов.

Пример:

```python
# Импортирование модуля PySimpleGUI для создания графического интерфейса
import PySimpleGUI as sg

# Определение макета интерфейса, который будет использоваться для окна
layout = [
    # Создание текстового элемента (метки), который будет спрашивать пользователя о любимом животном
    [sg.Text('Выберите ваше любимое животное:')],
    # Создание выпадающего списка с выборами 'Кошка', 'Собака', 'Кролик' и идентификатором 'animal_choice', включение событий для этого элемента
    [sg.Combo(['Кошка', 'Собака', 'Кролик'], key='animal_choice', enable_events=True)],
    # Создание кнопки 'OK', которая будет использоваться для подтверждения выбора пользователя
    [sg.Button('OK')]
]

# Создание окна с названием 'Мое окно' и заданным макетом
window = sg.Window('Мое окно', layout)

# Начало основного цикла программы, который будет выполняться до тех пор, пока пользователь не закроет окно
while True:
    # Чтение событий и значений из интерфейса
    event, values = window.read()
    # Проверка, не было ли закрыто окно пользователем через системные события (к примеру, нажатие на 'X')
    if event == sg.WINDOW_CLOSED:
        break # Прерывание цикла, закрытие окна
    # Проверка, была ли нажата кнопка 'OK'
    if event == 'OK':
        # Получение выбранного пользователем значения из выпадающего списка (используя ключ 'animal_choice')
        animal_choice = values['animal_choice']
        # Отображение всплывающего окна (попапа) с сообщением, включающим выбранный элемент
        sg.popup(f"Ваш выбор: {animal_choice}")

# Закрытие окна после завершения основного цикла программы
window.close()
```

> В этом коде создается простое окно с одним вопросом о любимом животном пользователя, выпадающим списком для выбора ответов, и кнопкой для подтверждения выбора. Пользователь выбирает животное из списка, нажимает 'OK', и появляется сообщение с выбранным вариантом.

Рассмотрим следующий пример:

```python
# Импортируется модуль PySimpleGUI под псевдонимом sg для создания графического интерфейса пользователя
import PySimpleGUI as sg

# Определяется макет (layout) для будущего окна. Макет определяет элементы интерфейса и их расположение.
layout = [
    # Эта строка содержит выпадающий список (DropDown) с тремя опциями: 'Вариант 1', 'Вариант 2', 'Вариант 3'.
    [sg.DropDown(['Вариант 1', 'Вариант 2', 'Вариант 3'])],
    # Эта строка включает элемент Output, который будет отображать принты и другие выводы, размер текстового поля 30x5 символов.
    [sg.Output(size=(30, 5))],
    # Эта строка добавляет кнопку 'Ввод', которая будет использоваться для подтверждения выбора из выпадающего списка.
    [sg.Button('Ввод')]
]

# Создается окно с заголовком 'Мое окно' и предварительно определенным макетом.
window = sg.Window('Мое окно', layout)

# Начинается цикл обработки событий окна.
while True:
    # Метод read() ожидает событие (например, нажатие кнопки) и возвращает имя события и словарь с данными элементов.
    event, values = window.read()  # event - название события, values - значения элементов формы
    # Если пользователь закроет окно, произойдет событие WINDOW_CLOSED и цикл прервется.
    if event == sg.WINDOW_CLOSED:
        break  # Прерывание главного цикла, что ведет к закрытию программы.
    # Если событие - это нажатие кнопки 'Ввод', тогда выполняется следующий блок кода.
    if event == 'Ввод':
        # Выводится значение, выбранное в выпадающем списке (значение первого элемента - values[0])
        print(values[0])  # Эта строка выводит выбранный вариант на элемент Output.

# При выходе из цикла окно закрывается.
window.close()  # Этот вызов удаляет окно из памяти и закрывает его.
```

> Код создает базовое графическое окно, которое содержит выпадающий список с тремя вариантами выбора, область вывода информации и кнопку 'Ввод'. Когда пользователь выбирает опцию из списка и нажимает 'Ввод', программа выводит выбранный вариант в области вывода. Программа ожидает действий пользователя до тех пор, пока он не закроет окно, после чего программа завершает работу и закрывает окно.

## 3.4 Метки и надписи

Метки (или надписи) обычно используются для отображения текста рядом с другими элементами GUI, такими как например ввод текста, чтобы пользователь понимал, что он должен ввести.


Пример кода, в котором создается окно с использованием PySimpleGUI с метками и полями для ввода:

```python
# Импорт библиотеки PySimpleGUI для создания графического пользовательского интерфейса (GUI)
import PySimpleGUI as sg

# Установка темы для графического интерфейса, "Dark Blue 3" - это название предустановленной темы
sg.theme('Dark Blue 3')

# Определение макета GUI — структуры интерфейса с виджетами
layout = [
    # Создание текстовой метки "Имя", рядом с которой расположено текстовое поле
    [sg.Text('Имя'), sg.InputText(key='name')],  # 'key' позволяет идентифицировать значение поля по ключу при чтении данных

    # Создание текстовой метки "Фамилия", рядом с которой расположено текстовое поле
    [sg.Text('Фамилия'), sg.InputText(key='surname')],  # Также, 'key' используется для идентификации ввода

    # Создание текстовой метки "Возраст", рядом виджет - список для выбора с числами от 1 до 100
    [sg.Text('Возраст'), sg.Spin(values=[i for i in range(1, 101)], initial_value=18, key='age')],
    # 'values' содержит список выбора, 'initial_value' - начальное значение

    # Создание кнопок "Отправить" и "Отмена" для управления формой
    [sg.Submit(button_text='Отправить'), sg.Cancel(button_text='Отмена')]  # 'button_text' изменяет название кнопки
]

# Создание окна с заголовком "Пример меток и надписей" и предопределенным макетом 'layout'
window = sg.Window('Пример меток и надписей', layout)

# Главный цикл программы для обработки событий интерфейса
while True:
    # Чтение событий и значений (нажатие кнопок, ввод данных и т.п.)
    event, values = window.read()

    # Если пользователь закрыл окно или нажал кнопку "Отмена"
    if event in (None, 'Отмена'):  # 'None' возвращается при закрытии окна "крестиком"
        break  # Прерывание главного цикла, для завершения программы

    # Обработка события нажатия на кнопку "Отправить"
    if event == 'Отправить':  # Проверка того что событие соответствует нажатию кнопки "Отправить"
        # В этом блоке можно выполнять действия с полученными данными, здесь они печатаются в консоль
        print(f"Имя: {values['name']}")  # Вывод в консоль введенного имени
        print(f"Фамилия: {values['surname']}")  # Вывод в консоль введенной фамилии
        print(f"Возраст: {values['age']}")  # Вывод в консоль выбранного возраста

# После выхода из цикла закрытие созданного окна
window.close()
```

> Этот пример создает окно с тремя метками и соответствующими элементами ввода данных. Каждому элементу ввода присвоен уникальный ключ для дальнейшей их идентификации при чтении данных. После нажатия на кнопку "Отправить" данные из полей ввода выведутся в консоль.

## 3.5 Списки и таблицы

Списки и таблицы позволяют отображать данные в упорядоченном виде. Они могут иметь различные настройки для управления отображением данных.

Пример:

```python
# Импортируем модуль PySimpleGUI для работы с графическим интерфейсом
import PySimpleGUI as sg

# Список списков, представляющий данные для таблицы
data = [
    [1, "Первый элемент", "Описание 1"],    # Первая строка таблицы
    [2, "Второй элемент", "Описание 2"],    # Вторая строка таблицы
    [3, "Третий элемент", "Описание 3"],    # Третья строка таблицы
]

# Список, содержащий заголовки столбцов для таблицы
headings = ["ID", "Название", "Описание"]

# Создаем список списков, где каждый внутренний список - это элементы интерфейса
layout = [
    [sg.Text("Таблица:")],  # Виджет текста для заголовка "Таблица"
    [sg.Table(values=data,                # Значения для отображения в таблице
              headings=headings,          # Заголовки столбцов таблицы
              max_col_width=25,           # Макс. ширина столбца
              auto_size_columns=True,     # Автоматический подбор ширины столбцов
              display_row_numbers=True,   # Отображение номеров строк
              justification='right',      # Выравнивание текста в ячейках по правому краю
              num_rows=5,                 # Количество строк, видимых в один момент времени
              key='-TABLE-',              # Уникальный ключ виджета (ищем по нему)
              row_height=35)],            # Высота строк в пикселях
    [sg.Text("Список:")],  # Виджет текста для заголовка "Список"
    # Виджет списка, который берет второй элемент (с индексом 1) из каждой строки таблицы data
    [sg.Listbox(values=[row[1] for row in data], size=(30, 6), key='-LISTBOX-')],
    [sg.Button("Выход")]   # Кнопка для выхода из программы
]

# Создаем окно с определенным названием и лейаутом
window = sg.Window("Пример таблицы и списка в PySimpleGUI", layout)

# Бесконечный цикл для обработки событий интерфейса
while True:
    event, values = window.read()  # Прочитать событие и значения виджетов окна
    # Если пользователь закрыл окно или нажал кнопку "Выход"
    if event == sg.WIN_CLOSED or event == 'Выход':
        break    # Выходим из цикла

# Закрываем окно после выхода из цикла обработки событий
window.close()
```

## 3.6 Чекбоксы и радиокнопки


Чекбоксы позволяют пользователю выбрать или снять определенное состояние, а радиокнопки - выбрать один из предопределенных вариантов.
Для использования чекбоксов и радиокнопок в PySimpleGUI необходимо использовать специальные виджеты: Checkbox и Radio.

```python
# Импорт модуля PySimpleGUI для создания GUI
import PySimpleGUI as sg

# Описание дизайна пользовательского интерфейса
layout = [
    # Добавление текстового элемента (вопрос)
    [sg.Text('Какой Ваш любимый цвет?')],
    # Группа радио-кнопок с одинаковым ключом "COLOR" для единственного выбора
    [sg.Radio('Красный', "COLOR", size=(10,1)),
     sg.Radio('Синий', "COLOR", size=(10,1)),
     sg.Radio('Зелёный', "COLOR", size=(10,1))],
    # Добавление еще одного текстового элемента
    [sg.Text('Выберите дополнительные опции:')],
    # Добавление чекбоксов для множественного не исключающего выбора
    [sg.Checkbox('Обводка', default=False, size=(10,1)),
     sg.Checkbox('Тень', default=False, size=(10,1))],
    # Добавление кнопок для отправки данных или отмены
    [sg.Submit(), sg.Cancel()]
]

# Создание окна с определенными параметрами, используя описанный выше дизайн
window = sg.Window('Чекбоксы и Радиокнопки', layout)

# Бесконечный цикл для обработки событий окна
while True:
    # Чтение событий и значений окна, блокируется пока не произойдет событие
    event, values = window.read()
    # Проверка на событие закрытия окна или нажатие кнопки "Cancel"
    if event in (None, 'Cancel'):
        break  # Выходим из цикла, закрываем окно
    # Проверка, если была нажата кнопка "Submit"
    elif event == 'Submit':
        # Установка значения по умолчанию для любимого цвета
        favorite_color = "Не выбрано"
        # Обновление переменной favorite_color, если выбрана определенная радио-кнопка
        if values[0]: favorite_color = 'Красный'
        if values[1]: favorite_color = 'Синий'
        if values[2]: favorite_color = 'Зелёный'

        # Проверка выбранных чекбоксов для формирования строки с доп. опциями
        options = "Обводка" if values[3] else ""
        # Если выбрано несколько опций, добавляем соединитель "и"
        options += " и " if values[3] and values[4] else ""
        # Добавление в строку опции "Тень", если чекбокс выбран
        options += "Тень" if values[4] else ""

        # Выводимый пользователю попап с информацией о выборе
        sg.Popup(f'Ваш любимый цвет: {favorite_color}\nДополнительные опции: {options}')

# Закрытие окна после выхода из цикла
window.close()
```

В этом примере есть следующие ключевые моменты:

- `sg.Radio` создает радиокнопки. Параметр "COLOR" гарантирует, что они объединены в одну группу, где можно выбрать только один вариант.
- `sg.Checkbox` создает чекбоксы; параметры `default` и `size` настраивают начальное состояние и размеры.
- В бесконечном цикле программа ждёт действий пользователя. При нажатии кнопки "Submit", программа считывает значения введенные пользователем.
- `window.read()` возвращает два значения: событие (нажатая кнопка) и словарь значений, где ключи - это индексы элементов (для чекбоксов и радио-кнопок).
- `sg.Popup` отображает всплывающее окно с сообщением пользователю.

### Подробнее разберем универсальный метод popup_get_file():

Метод popup_get_file в PySimpleGUI — это удобная функция для отображения стандартного диалогового окна выбора файла, которая позволяет пользователю выбрать файл для открытия. Это часть более широкого набора функций "popup" в PySimpleGUI, которые представляют собой простой способ создания наиболее распространенных элементов интерфейса без необходимости ручной настройки элементов окна.

Другой пример использования popup_get_file():


```python
# Импортируем модуль PySimpleGUI для работы с графическим интерфейсом пользователя
import PySimpleGUI as sg
# Импортируем модуль os для работы с операционной системой
import os

def copy_file(file_name):
    # Преобразование пути файла для совместимости с текущей операционной системой
    file_name = file_name.replace('/', '\\')
    # Формирование имени для копии файла
    copy_name = f"{os.path.splitext(file_name)[0]}_copy{os.path.splitext(file_name)[1]}"
    # Копирование файла с добавлением суффикса "_copy" к имени файла
    os.system(f"copy {file_name} {copy_name}")
    # Возврат полного имени копии файла
    return copy_name

# Определение макета графического интерфейса пользователя
layout = [
    [sg.Button('Сохранить как')], # Добавляем кнопку "Сохранить как"
    [sg.Button('Выход')]          # Добавляем кнопку "Выход"
]

# Создание окна с заголовком 'Мое окно' и макетом layout
window = sg.Window('Мое окно', layout)

# Цикл обработки событий в графическом интерфейсе
while True:
    event, values = window.read()  # Считывание событий и значений из элементов интерфейса

    # Обработка событий
    if event == sg.WINDOW_CLOSED or event == 'Выход':
        break  # Завершение программы, если окно закрыто или нажата кнопка "Выход"
    elif event == 'Сохранить как':
        # Открытие диалогового окна для сохранения файла
        file_name = sg.popup_get_file('Выберите файл для сохранения', save_as=True, no_window=True)

        if file_name and file_name != '':
            copy_name = copy_file(file_name)
            sg.popup(f"Файл сохранен как: {copy_name}")

# Закрытие окна после выхода из цикла
window.close()
```

> При нажатии на "Сохранить как"  вызывается метод popup_get_file(), после чего пользователю будет предложен выбор файла для копирования через диалоговое окно. После выбора файла функция `copy_file` копирует указанный файл, а затем пользователю показывается всплывающее окно с путем к скопированному файлу. Если пользователь нажимает "Выход" или закрывает окно, программа завершается.

Опции метода `popup_get_file`:

-   `message` - Текст, отображаемый внутри диалогового окна. Например, "Please choose a file."
-   `default_path` - Путь к каталогу, который будет открыт по умолчанию когда диалоговое окно запустится.
-   `default_extension` - Расширение файла, выбранное по умолчанию.
-   `save_as` - Если True, диалог будет работать как "Сохранить как...". По умолчанию False.
-   `file_types` - Фильтры файлов, отображаемые в диалоговом окне, в формате списка кортежей. Например, `(("Text Files", "*.txt"),("All Files", "*.*"))`.
-   `no_window` **- Если True, всплывающее окно не будет блокировать остальные окна вашего приложения после закрытия диалогового окна. По умолчанию False.**
-   `size` (width, height) - размеры всплывающего окна.
-   `location` (x, y) - расположение всплывающего окна на экране.
-   `title` - Заголовок всплывающего окна.
-   `button_color` - Цвет кнопки.
-   `background_color` - Цвет фона.
-   `text_color` - Цвет текста.
-   `icon` - Путь к файлу с иконкой для всплывающего окна.

В PySimpleGUI также есть связанные функции для других стандартных диалогов:

-   `popup_get_folder` - функция для выбора папки.
-   `popup_get_text` - функция для ввода текста.
-   `popup_an_error` - функция для отображения сообщения об ошибке.
-   `popup` - базовая функция для отображения всплывающего окна с заданным сообщением.


## 4.3. Использование тем оформления

PySimpleGUI поддерживает различные темы оформления, которые позволяют изменить внешний вид окон. Для изменения темы оформления можно использовать метод sg.theme(). Например, чтобы установить тему оформления "DarkAmber", можно сделать следующее:

```python
import PySimpleGUI as sg  # Импорт библиотеки PySimpleGUI с псевдонимом sg

sg.theme_previewer()  # Отображает предварительный просмотр тем оформления, доступных в PySimpleGUI

sg.theme("DarkAmber")  # Устанавливает тему оформления для всех элементов GUI

# Описание макета окна GUI
layout = [
    [sg.Text("Hello, world!")],  # Добавляет виджет текста с сообщением "Hello, world!"
    [sg.Button("OK")]           # Добавляет кнопку с текстом "OK"
]

# Создание объекта Window, который будет показывать все наши виджеты, с заголовком "My Window" и макетом layout
window = sg.Window("My Window", layout)

# Бесконечный цикл для считывания событий и значений из элементов GUI
while True:
    event, values = window.read()  # Читает события, происходящие в окне, и сохраняет их в переменных event и values
    if event in (sg.WINDOW_CLOSED, "OK"):  # Проверяет, были ли закрыто окно или нажата кнопка "OK".
        break  # Выходит из цикла, если было обнаружено событие закрытия окна или кнопка "OK" была нажата

window.close()  # Закрывает окно
```
